```python
prep = PreProcess(df_sea)
prep.drop_columns()
prep.preprocess_df()
prep.create_parking_index()
cols = ['bathrooms', 'bedrooms', 'finishedsqft', 'yearbuilt']
prep.normalize_columns(cols)
prep.df[['trans_score', 'walkscore_score']] = prep.df[['trans_score', 'walkscore_score']] / 100.0
prep.df['school_index'] = prep.df['school_index'] / 10.0
prep.df.head()
```

```python
prep2 = PreProcess(df_sf)
prep2.drop_columns()
prep2.preprocess_df()
prep2.create_parking_index()
cols = ['bathrooms', 'bedrooms', 'finishedsqft', 'yearbuilt']
prep2.normalize_columns(cols)
prep2.df[['trans_score', 'walkscore_score']] = prep2.df[['trans_score', 'walkscore_score']] / 100.0
prep2.df.head()
```
