```python
def extract_address_and_zpid(list_html_content):
    '''
    Input: a list of html_source for different web_pages 
    
    Output: a pandas dataframe
    '''
    homes = {}

    for html_doc in html_content:
        soup = BS(html_doc, 'html.parser')
        indices = [m.start() for m in re.finditer('_zpid', html_doc)]

        zids = []

        for index in indices:
            string = html_doc[index - 10: index]
            if "/" in string:
                string = string[string.find('/') + 1 : index]
            if string not in zids:
                zids.append(string)

        # builds a list of links from the html_source 

        links =  [res['href'] for res in soup.find_all('a', attrs={"href": True})]

        # Takes links and zids and returns the address of a specific zid
        for link in links:
            for zid in zids:
                if (zid in link) and ('homedetails' in link):
                    if zid not in homes:
                        start_index = link.find('/', 1)
                        stop_index = link.find('/', start_index + 1)
                        homes[zid] = link[start_index + 1 : stop_index]
        
        return pd.DataFrame({'zid': homes.keys(), 'address': homes.values()}, index=range(len(homes)))
```

```python
def extract_image_urls(html_source, num_images):
    
    soup = BS(html_source, 'html.parser')
    images = soup.find_all('img')
    
    image_urls = []

    for img in images:
        try:
            image = img['src']
        except:
            try:
                image = img['href']
            except:
                pass
        if image[-4:] == '.jpg':
            image_urls.append(image)
        if len(image_urls) == num_images:
            break
    
    return image_urls
```

```python
# wait_between_sprints is a list consisting of [min_wait, max_wait]
# e.g. [30, 90] will allow the function to choose a random sleep time between
# 30 and 90 seconds

def scrape_a_city(df, sprint_length, wait_between_sprints):
    
    count = 0
    top_directory = os.getcwd()
    
    session = requesocks.session()
    session.proxies = {'http': 'socks5://127.0.0.1:9150',
                        'https': 'socks5://127.0.0.1:9150'}
    
    for row in df.iterrows():

        if count % 20 == 0:
            time.sleep(90)
        if count % sprint_length == 0:
            sleep_time = random.randint(wait_between_sprints[0], wait_between_sprints[1])
            time.sleep(sleep_time)
            
        # Iterate through the address and zid for each the dataframe
        address = row[1][0]
        zid = row[1][1]
        
        # Make a directory for each address to put all of the scraped info
        # and then cd into new_directory to write all scraped images and html to it
        
        new_directory = address + '_' + zid
        
        os.mkdir(new_directory)
        os.chdir(new_directory)
        
        # Construct an html query for each address
        url = "http://www.zillow.com/homes/for_sale/" + address + '/' + zid + '_zpid/'

        try:
            
            response = session.get(url)
            html_source = response.content
            
            # Write the html_source to a txt file for parsing later
            print os.getcwd()
            with open('html_source.txt', 'w') as f:
                f.write(html_source)

            image_urls = extract_image_urls(html_source, 15)
            get_images(image_urls)

        except:
            pass
        
        # cd back into the top directroy to begin scraping the next address
        os.chdir(top_directory)
        
        count += 1
```
